```python
# Read csv File
import pandas as pd
df=pd.read_csv("Students_Marks_SY1_SY2_SY3.csv")
df
```

```python
# Add a column Name is : Total
df['Total'] = df[['Math', 'Physics', 'Chemistry', 'English', 'Computer']].sum(axis=1)
df
```

```python
# Add a column name : Percentage
df['Percentage'] = (df['Total'] / 500) * 100
df
```

```python
# import liberary numpy as np
import numpy as np
```

```python
# Using np.where function & add Grade for using percentage column
df["grade"]=np.where(df["Percentage"]>90,"A",
            (np.where(df["Percentage"]>80,"B",
            (np.where(df["Percentage"]>70,"C",
            (np.where(df["Percentage"]>60,"D",
            (np.where(df["Percentage"]>50,"E",
            (np.where(df["Percentage"]>40,"F","Fail")))))))))))
df
```

```python
# Find a mean_score
mean_scores = df["Total"].mean()
print(mean_scores)
```

```python
# Bar plot ::::Average Marks by Subjects

import matplotlib.pyplot as plt
import numpy as np

subjects = ['Math', 'Physics', 'Chemistry', 'English', 'Computer']
avg_marks = df[subjects].mean()

plt.figure(figsize=(6,5))
plt.bar(subjects, avg_marks, color='pink')
plt.title("Average Marks by Subject")
plt.ylabel("Average Marks")
plt.xlabel("Subjects")
```

```python
# Heatmap:::::::::Subject wise Corelation Metrix

import seaborn as sns

plt.figure(figsize=(8,6))
corr_matrix = df[['Math', 'Physics', 'Chemistry', 'English', 'Computer']].corr()

sns.heatmap(corr_matrix, annot=True, cmap='cubehelix', linewidths=0.6)
plt.title("Subject Wise Marks Correlation")
plt.show()
```

```python
#Topper pr Department:::::::::::
df['Percentage'] = df[subjects].sum(axis=1) / len(subjects)
topper = df.loc[df.groupby('department')['Percentage'].idxmax()]
counts = topper['department'].value_counts()

plt.figure(figsize=(8,5))
plt.pie(counts, labels=counts.index, autopct='%1.1f%%', startangle=50, colors=['#8e44ad', '#16a085', '#f39c12'], wedgeprops={'width':0.5})
plt.title("Topper per Department")
plt.tight_layout()
plt.show()
```

```python
# KNN Model::::::::::::Using Machine Learning Prediction

import pandas as pd
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
from sklearn.preprocessing import StandardScaler

f=pd.read_csv("Students_Marks_SY1_SY2_SY3.csv")
df=pd.DataFrame(f)
df
```

```python
x = df[["Math","Physics","Chemistry","English","Computer"]]
y=df["department"]

x=pd.get_dummies(x)

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
model=KNeighborsClassifier(n_neighbors=8)

model.fit(x_train,y_train)
pre=model.predict(x_test)

acc=accuracy_score(y_test,pre)*100
print(f"accuracy is {acc}")

con=confusion_matrix(y_test,pre)
print(con)

cl=classification_report(y_test,pre)
print(f"classification_report is {cl}")
```

```python
x = df[["Math","Physics","Chemistry","English","Computer"]]
y=df["Division"]

x=pd.get_dummies(x)

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
model=KNeighborsClassifier(n_neighbors=8)

model.fit(x_train,y_train)
pre=model.predict(x_test)

acc=accuracy_score(y_test,pre)*100
print(f"accuracy is {acc}")

con=confusion_matrix(y_test,pre)
print(con)

cl=classification_report(y_test,pre)
print(f"classification_report is {cl}")
```
